```python
import numpy as np # linear algebra
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
movies = pd.read_csv('tmdb_5000_credits.csv')
credits = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','vote_count','vote_average']]
```

```python
movies['votes']= movies.vote_count * movies.vote_average
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','votes']]
                 
```

```python
movies.head()
```

```python
popularity_threshold = 45000
votes = movies.query('votes >= @popularity_threshold')
```

```python
votes.head()
```

```python
movie_features_df=votes.pivot_table(index='title',columns='movie_id',values='votes').fillna(0)
movie_features_df.head()
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.shape
```

```python
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
movie_features_df.head()
```

```python
def recommend(movie):
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i])) 
```

```python
recommend('Avatar')
```
